1 -  Carregando o arquivo Excel

```python
import pandas as pd

arquivo = pd.read_excel("cursinho.xlsx")
display(arquivo)
```

2 - Removendo Linhas e Colunas vazias

```python
#removendo coluna vazia
arquivo.dropna(how='all', axis=1, inplace=True) 

#transformando a primeira linha em cabeçalho
arquivo.columns = arquivo.iloc[0]
arquivo = arquivo[1:]
display(arquivo)
```

3- Total de Faltas durante a semana

```python
total_faltas = arquivo["Alunos com falta"].sum()
print(f'Total de faltas na semana: {total_faltas} faltas')
```

4 - Desenhando gráfico de pizza com a quantidade de faltas por Matéria

```python
import matplotlib.pyplot as plt

materia = arquivo["Matérias"]
faltas = arquivo["Alunos com falta"]

#desenhando o gráfico

plt.title("Faltas por Matéria na semana")
plt.pie(faltas, labels=materia, autopct="%1.1f%%")
plt.show()
```

5 - Variação de faltas durante a semana

```python
diaSemana = arquivo["Dias da semana"]

#desenhando o gráfico

plt.title("Faltas durante a semana")
plt.plot(diaSemana,faltas, marker='D', ms=5, linestyle=':', color='red')
plt.grid()
plt.show()
```

6 - Variação de Presenças durante a semana

```python
presencas = arquivo["Alunos presentes"]

plt.title("Faltas vs Presenças durante a semana")
plt.plot(diaSemana,presencas, marker='D', ms=5, linestyle='-', color='blue')
plt.grid()
plt.show()
```

7- Média de Presenças x Média de faltas

```python
#calculando as médias
media_presencas = presencas.mean()
media_faltas = faltas.mean()

medias = [media_presencas,media_faltas]
label = [f'Presenças ({media_presencas:.0f})', f'Faltas ({media_faltas:.0f})']

#gerando o gráfico
plt.title("Média de Presença vs Faltas durante a semana")
plt.pie(medias, labels=label, autopct="%1.0f%%")
plt.show()
```
